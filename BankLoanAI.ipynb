```python
#Back-End
import random
import pandas as pd
rand = random.randint(1, 200000)
df_boa_loan_data = pd.read_csv("/Users/brandonelijahisbell/Downloads/aloan-algorithm/boa-loan-data.csv")
boa_raw_data = df_boa_loan_data["Data"].tolist()[rand:rand+20000]

df_jpm_loan_data = pd.read_csv("/Users/brandonelijahisbell/Downloads/aloan-algorithm/jpm-loan-data.csv")
jpm_raw_data = df_jpm_loan_data["Data"].tolist()[rand:rand+20000]

df_pnc_loan_data = pd.read_csv("/Users/brandonelijahisbell/Downloads/aloan-algorithm/pnc-loan-data.csv")
pnc_raw_data = df_pnc_loan_data["Data"].tolist()[rand:rand+20000]

df_ctz_loan_data = pd.read_csv("/Users/brandonelijahisbell/Downloads/aloan-algorithm/ctz-loan-data.csv")
ctz_raw_data = df_ctz_loan_data["Data"].tolist()[rand:rand+20000]

df_tru_loan_data = pd.read_csv("/Users/brandonelijahisbell/Downloads/aloan-algorithm/tru-loan-data.csv")
tru_raw_data = df_tru_loan_data["Data"].tolist()[rand:rand+20000]

raw_datas = []
raw_datas.append(boa_raw_data)
raw_datas.append(jpm_raw_data)
raw_datas.append(pnc_raw_data)
raw_datas.append(ctz_raw_data)
raw_datas.append(tru_raw_data)

def preprocessing():
    training_datas_normalized = []
    for raw_data in raw_datas:
        training_data_not_normalized = []
        for raw_data_string in raw_data:
            raw_data_list_unordered = raw_data_string.split("|")
            raw_data_ordered = []
            # County Code
            raw_data_ordered.append(raw_data_list_unordered[10])
            # Income
            raw_data_ordered.append(raw_data_list_unordered[44])
            # Age
            raw_data_ordered.append(raw_data_list_unordered[40])
            # Gender
            raw_data_ordered.append(raw_data_list_unordered[36])
            # Race
            raw_data_ordered.append(raw_data_list_unordered[24])
            # Loan Amount
            raw_data_ordered.append(raw_data_list_unordered[7])
            # Debt-Income Ratio
            raw_data_ordered.append(raw_data_list_unordered[62])
            # Bank
            raw_data_ordered.append(raw_datas.index(raw_data))
            training_data_not_normalized.append(raw_data_ordered)

        training_data_normalized = []
        for raw_data_not_normalized in training_data_not_normalized:
            raw_data_normalized = []
            # County Code
            if raw_data_not_normalized[0] == "NA":
                raw_data_normalized.append(0)
            else:
                raw_data_normalized.append(raw_data_not_normalized[0])
            # Income
            if raw_data_not_normalized[1] == "NA":
                raw_data_normalized.append(0)
            else:
                raw_data_normalized.append(int(raw_data_not_normalized[1]))
            # Age
                # 0 = 8888, NA
                # 1 = <25
                # 2 = 25-34
                # 3 = 35-44
                # 4 = 45-54
                # 5 = 55-64
                # 6 = 65-74
                # 7 = >74
            if raw_data_not_normalized[2] == "8888":
                raw_data_normalized.append(0)
            elif raw_data_not_normalized[2] == "<25":
                raw_data_normalized.append(1)
            elif raw_data_not_normalized[2] == ">74":
                raw_data_normalized.append(7)
            else:
                raw_data_normalized.append(int(list(raw_data_not_normalized[2])[0]))
            # Gender
                # 0 = NA
                # 1 = Male
                # 2 = Female
                # 3 = Other
                # 4 = Not Reported
            if raw_data_not_normalized[3] == "6":
                raw_data_normalized.append(0)
            else:
                raw_data_normalized.append(int(raw_data_not_normalized[3]))
            # Race
                # 0 = NA
                # 1 = American Indian or Alaskan
                # 2 = Asian
                # 3 = Black or African American
                # 4 = Native Hawaiian or Pacific Islander
                # 5 = White
                # 6 = Not Reported
            if raw_data_not_normalized[4] == "7":
                raw_data_normalized.append(0)
            else:
                raw_data_normalized.append(int(list(raw_data_not_normalized[3])[0]))
            # Loan Amount
            if raw_data_not_normalized[5] == "NA":
                raw_data_normalized.append(0)
            else:
                raw_data_normalized.append(int(raw_data_not_normalized[5]))
            # Debt-Income Ratio
                # 0 = NA
                # 1 = <20%
                # 2 = 20%-<30%
                # 3 = 30%-<40%
                # 4 = 40%-<50%
                # 5 = 50%-60%
                # 6 = >60%
                # 7 = Exempt
            if raw_data_not_normalized[6] == "NA":
                raw_data_normalized.append(0)
            elif raw_data_not_normalized[6] == "<20%":
                raw_data_normalized.append(1)
            elif raw_data_not_normalized[6] == ">60%":
                raw_data_normalized.append(6)
            elif raw_data_not_normalized[6] == "Exempt":
                raw_data_normalized.append(7)
            else:
                raw_data_normalized.append(int(list(raw_data_not_normalized[5])[0]))
            raw_data_normalized.append(raw_data_not_normalized[7])
            training_data_normalized.append(raw_data_normalized)

        training_datas_normalized.append(training_data_normalized)

    training_datas_normalized
    df_training_datas = []

    for df_training_data in training_datas_normalized:
        df_training_data = pd.DataFrame(df_training_data, columns=["County Code", "Income", "Age", "Gender", "Race", "Loan", "Ratio", "Bank"])
        df_training_datas.append(df_training_data)

    df_training_datas = pd.concat(df_training_datas)

    df_training_datas["County Code"] = df_training_datas["County Code"].astype("object")
    df_training_datas["Age"] = df_training_datas["Age"].astype("object")
    df_training_datas["Gender"] = df_training_datas["Gender"].astype("object")
    df_training_datas["Race"] = df_training_datas["Race"].astype("object")
    df_training_datas["Ratio"] = df_training_datas["Ratio"].astype("object")
    df_training_datas["Bank"] = df_training_datas["Bank"].astype("object")
    return df_training_datas
def train_and_test(zip_code, income, age, gender_num, race_num, loan_amount, debt_income_num, bank_num):
    df_training_datas = preprocessing()
    from sklearn.datasets import load_iris
    from sklearn.linear_model import LogisticRegression
    from sklearn.model_selection import train_test_split

    training_data_answers = []

    for raw_data_string in boa_raw_data:
        raw_data_list_unordered = raw_data_string.split("|")
        raw_data_ordered = []
        if raw_data_list_unordered[51] == "10":
            training_data_answers.append(1)
        else:
            training_data_answers.append(0)
            
    for raw_data_string in jpm_raw_data:
        raw_data_list_unordered = raw_data_string.split("|")
        raw_data_ordered = []
        if raw_data_list_unordered[51] == "10":
            training_data_answers.append(1)
        else:
            training_data_answers.append(0)

    for raw_data_string in pnc_raw_data:
        raw_data_list_unordered = raw_data_string.split("|")
        raw_data_ordered = []
        if raw_data_list_unordered[51] == "10":
            training_data_answers.append(1)
        else:
            training_data_answers.append(0)

    for raw_data_string in ctz_raw_data:
        raw_data_list_unordered = raw_data_string.split("|")
        raw_data_ordered = []
        if raw_data_list_unordered[51] == "10":
            training_data_answers.append(1)
        else:
            training_data_answers.append(0)

    for raw_data_string in tru_raw_data:
        raw_data_list_unordered = raw_data_string.split("|")
        raw_data_ordered = []
        if raw_data_list_unordered[51] == "10":
            training_data_answers.append(1)
        else:
            training_data_answers.append(0)

    X = df_training_datas
    y = training_data_answers

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=57)

    from sklearn.ensemble import AdaBoostClassifier
    from sklearn.datasets import make_classification
    from sklearn.metrics import classification_report

    abc = AdaBoostClassifier(n_estimators=50,
                            learning_rate=1)

    model = abc.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    input = [zip_code, income, age, gender_num, race_num, loan_amount, debt_income_num, bank_num]
    df_input = pd.DataFrame(input).transpose()

    prediction = model.predict(df_input)
    return int(prediction)



# County Code, Income, Age, Gender, Race, Loan Amount, Debt-Income Ratio, Bank
def run_model(zip_code, income, age, gender_num, race_num, loan_amount, debt_income_num, bank_num):
    return train_and_test(zip_code, income, age, gender_num, race_num, loan_amount, debt_income_num, bank_num)

def vars_to_numerical(gender, age, race, income, debt, bank):
    age = int(age)
    income = int(income)
    debt = int(debt)
    gender = str(gender)
    # Input Validation
    if not isinstance(age, (int, float)) or not isinstance(income, (int, float)) or not isinstance(debt, (int, float)):
        raise ValueError("Age, income, and debt must be numeric values.")
    gender_mapping = {
        "NA": -1,
        "Male": 0,
        "Female": 1,
        "Other": 2
    }

    race_mapping = {
        "NA": -1,
        "American Indian or Alaskan": 0,
        "Asian": 1,
        "Black or African American": 2,
        "Native Hawaiian or Pacific Islander": 3,
        "White": 4
    }

    age_mapping = {
        (1, 25): 1,
        (25, 35): 2,
        (35, 45): 3,
        (45, 55): 4,
        (55, 65): 5,
        (65, 75): 6,
        (75, float('inf')): 7
    }

    # Gender
    gender_num = gender_mapping.get(gender, -1)

    # Race
    race_num = race_mapping.get(race, -1)

    # Age
    age_num = -1  # Default value for unknown age
    for age_range, num in age_mapping.items():
        if age_range[0] <= age < age_range[1]:
            age_num = num
            break

    # Debt-Income Ratio
    debt_income_ratio = debt / income

    debt_income_mapping = {
        (0, 0.2): 1,
        (0.2, 0.3): 2,
        (0.3, 0.4): 3,
        (0.4, 0.5): 4,
        (0.5, 0.6): 5,
        (0.6, float('inf')): 6
    }

    debt_income_num = -1  # Default value for unknown debt-income ratio
    for ratio_range, num in debt_income_mapping.items():
        if ratio_range[0] <= debt_income_ratio < ratio_range[1]:
            debt_income_num = num
            break

    # Bank
    bank_mapping = {
        "Bank of America": 0,
        "JP Morgan": 1,
        "PNC": 2,
        "Citizens Bank": 3,
        "Truist": 4
    }

    bank_num = bank_mapping.get(bank, -1)

    return gender_num, race_num, age_num, debt_income_num, bank_num

#Front-End
from taipy.gui import Gui, notify
import random

text = "Your name here"
income = 50000
debt = 0
race = "Your race here"
age = 16
gender = "Your gender here"
zip_code = 99999
loan_amount = 0
debt_income_ratio = -1
loan_value_ratio = 0
bank = "Your Bank Here"
content = "https://upload.wikimedia.org/wikipedia/commons/9/90/Magic8ball.jpg"

def on_button_action(state):
    gender_num, race_num, age_num, debt_income_num, bank_num = vars_to_numerical(state.gender, state.age, state.race, state.income, state.debt, state.bank)
    result = "Granted" if run_model(state.zip_code, state.income, age_num, gender_num, race_num, state.loan_amount, debt_income_num, bank_num)==1 else "Not Granted"
    notify(state, 'info', f'Your loan result is: {result}', system_notification=True)
    

page = """
# Think you can secure a loan? 
#
Input your information below
#
Name: <|{text}|>

<|{text}|input|>

*Demographic Information*


Age: <|{age}|>

<|{age}|slider|min=16|max=120|>

Race: <|{race}|selector|lov=White;Black or African American;Asian;Native Hawaiian or Pacific Islander;American Indian or Alaskan;NA|dropdown|>

Gender: <|{gender}|selector|lov=Male;Female;Other;Prefer not to Answer|dropdown|>

*Personal Info*


Yearly Income: <|{income}|>

<|{income}|slider|min=0|max=1000000|>

Yearly Debt: <|{debt}|>

<|{debt}|input|>

Zip Code: <|{zip_code}|>

<|{zip_code}|input|>

*Loan Info*

Bank: <|{bank}|selector|lov=Bank of America;JP Morgan;PNC;Citizens Bank;Truist|dropdown|>

Loan Amount: <|{loan_amount}|>

<|{loan_amount}|slider|min=0|max=100000|>

<|Submit Loan Request|button|on_action=on_button_action|>

<|{content}|image|label=Submit a Loan Request|on_action=on_button_action|>
"""

Gui(page).run()
```
